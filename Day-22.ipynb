```python
#Tree Implementation

class Treenode():
    def __init__(self, value):
        self.value = value
        self.children = []

ceo = Treenode("CEO")
cfo = Treenode("CFO")
cto = Treenode("CTO")
coo = Treenode("COO")
qa = Treenode("QA")
dev = Treenode("DEV")
ops = Treenode("OPS")


ceo.children = [cfo, cto, coo]
cto.children = [qa, dev]
coo.children = [ops]

def traversal(node, level=0):
    print("      " * level + node.value)
    for child in node.children:
        traversal(child, level + 1)

traversal(ceo)


```

```python
# Binary Tree 
class node():
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return node(value)
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root
def print_tree(node,level=0):
    if node is not None:
        print_tree(node.right,level+1)
        print("       "*level,node.value)
        print_tree(node.left,level+1)

values = [20, 10, 30, 40, 25, 18, 5]

root = None
for val in values:
    root = insert(root, val)
print_tree(root)    
```

```python
# Binary Tree 
class node():
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return node(value)
    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)
    return root
#INORDER
def inorder(root):
    if root:
        inorder(root.left)
        print(root.value,end=" ")
        inorder(root.right)

#PREORDER
def preorder(root):
    if root:
        print(root.value,end=" ")
        preorder(root.left)
        preorder(root.right)        

#POSTORDER
def postorder(root):
    if root:
        preorder(root.left)
        preorder(root.right) 
        print(root.value,end=" ")       


values = [20, 10, 30, 40, 25, 18, 5]

root = None
for val in values:
    root = insert(root, val)

print("Inorder Traversal")
inorder(root)
print("\n")
print("Preorder Traversal")
preorder(root)
print("\n")
print("Postorder Traversal")
postorder(root)
print("\n") 
print_tree(root)   
```

```python
#height=max(height of left,height of right)+1
def height(root):
    if not root:
        return -1
    return 1+max(height(root.left),height(root.right))
print(height(root))
```

```python
def count_ln(root):
    if not root:
        return 0
    if not root.left and not root.right:
        return 1
    return count_ln(root.left)+ count_ln(root.right)
print(count_ln(root))
```

```python
# check the tree is balanced or not
def balance(root):
    if not root:
        return 0
    

    
    if balance(root.left)==balance(root.right):
        return True
    return False
print(balance(root))
```

```python



# Assignment 5  - 7/8/2025

# 1. Organizational Hierarchy (Tree View)
# Scenario:
# A company’s reporting structure:
#           CEO
#          /   \
#     Manager1 Manager2
#      /             \
#   Employee1     Employee2
# Why Binary Tree?
# •	Each person can report to one manager (parent), and may have up to 2 direct subordinates.
# Use-case:
# •	Traversals (level order, in-order) can be used to generate org charts or send emails level-wise.



class Node():
    def __init__(self, val): 
        self.val = val
        self.left = None
        self.right = None


root = Node("CEO")
root.left = Node("Manager1")

root.right = Node("Manager2")
root.left.left = Node("Employee1")
root.right.right = Node("Employee2")


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print("       " * level + str(node.val))
        print_tree(node.left, level + 1)


print_tree(root)
```

```python
# 2. Tournament Bracket System
# Scenario:
# Any knockout-style tournament:
#        Final
#       /     \
#   Semi1     Semi2
#    /  \     /  \
#  Q1  Q2   Q3  Q4
# Why Binary Tree?
# •	Each match has two players (left/right).
# •	The winner moves up as the parent node (next match).
# Use-case:
# •	Winner determination, path to final, etc.


class Node():
    def __init__(self, val):   
        self.val = val
        self.left = None
        self.right = None

q1 = Node("Q1")
q2 = Node("Q2")
q3 = Node("Q3")
q4 = Node("Q4")

semi1 = Node("Semi1")
semi1.left = q1
semi1.right = q2

semi2 = Node("Semi2")
semi2.left = q3
semi2.right = q4
final = Node("Final")
final.left = semi1
final.right = semi2
def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)
print_tree(final)
print("\n\nInorder Traversal:")
inorder(final)

print("\n\nPreorder Traversal:")
preorder(final)

print("\n\nPostorder Traversal:")
postorder(final)
```

```python
#3. Expression Evaluation in Compilers (Expression Tree)
#Scenario:
#Evaluate math expression:
#(3 + 5) * (2 - 1)
#Tree:
 #      *
 #    /   \
 #   +     -
 #  / \   / \
#  3  5  2  1
#Why Binary Tree?
#•	Each operator has 2 operands → left/right children


class Node:
    def __init__(self, val): 
        self.val = val
        self.left = None
        self.right = None

root = Node("*")
root.left = Node("+")
root.right = Node("-")

root.left.left = Node("3")
root.left.right = Node("5")
root.right.left = Node("2")
root.right.right = Node("1")

def inorder(root):
    if root:
        if root.left or root.right:
            print("(", end="")
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
        if root.left or root.right:
            print(")", end="")
def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def evaluate(root):
    if not root.left and not root.right:
        return int(root.val)
    l = evaluate(root.left)
    r = evaluate(root.right)
    if root.val == '+': return l + r
    if root.val == '-': return l - r
    if root.val == '*': return l * r
    if root.val == '/': return l // r

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)

print("Tree Structure:")
print_tree(root)

print("\nInorder Expression:")
inorder(root)

print("\n\nPreorder Expression:")
preorder(root)

print("\n\nPostorder Expression:")
postorder(root)

print("\n\nEvaluated Result:")
print(evaluate(root))
```
